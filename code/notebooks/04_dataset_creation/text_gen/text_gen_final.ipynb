```python
import json

# SETUP — Update these paths
input_path = "/content/drive/MyDrive/biomedical_text_generation/data/enriched/abstracts_to_text.json"
output_jsonl_path = "/content/drive/MyDrive/biomedical_text_generation/data/final_datasets/title_text_gen_ready.jsonl"

# Load JSON
with open(input_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# Convert to list if it's a single dictionary
if isinstance(data, dict):
    data = [data]

# Extract input/target pairs
records = []
skipped = 0

for item in data:
    title = item.get('title', '').strip()
    abstract = item.get('abstract', '').strip()

    # Skip bad entries
    if not title or not abstract:
        skipped += 1
        continue

    records.append({
        'input': title,
        'target': abstract
    })

# Save to JSONL
with open(output_jsonl_path, 'w', encoding='utf-8') as f:
    for r in records:
        json.dump(r, f, ensure_ascii=False)
        f.write('\n')

# Output confirmation
print(f"JSONL saved to: {output_jsonl_path}")
print(f"Total records: {len(records)}")
print(f"Skipped records: {skipped}")

print("\nSample entry:")
print(json.dumps(records[0], indent=2, ensure_ascii=False))
```
