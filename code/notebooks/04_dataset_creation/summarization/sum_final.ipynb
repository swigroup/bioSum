```python
import json
import os
import pandas as pd

# SETUP — Update these paths
input_path = "/content/drive/MyDrive/biomedical_text_generation/data/enriched/abstracts_to_text.json"   # JSON file path
output_jsonl_path = "/content/drive/MyDrive/biomedical_text_generation/data/final_datasets/summarization_ready.jsonl"

# Choose target style: 'matched'(no title included) or 'title_matched'(title included)
target_style = 'title_matched'  # or 'title_matched'

# Load JSON
with open(input_path, 'r') as f:
    data = json.load(f)

# Convert to list if it's a single dictionary (i.e., one example only)
if isinstance(data, dict):
    data = [data]

#  Extract input/target pairs
records = []
for item in data:
    #input_text = "summarize: " + item.get('abstract', '').strip() # Only for T5 models
    input_text = item.get('abstract', '').strip()


    if target_style == 'matched':
        target_text = item.get('matched_text', '').strip()
    elif target_style == 'title_matched':
        title = item.get('title', '').strip()
        matched = item.get('matched_text', '').strip()
        if title.endswith('.'):
            target_text = f"{title} {matched}"
        else:
            target_text = f"{title}. {matched}"

    else:
        raise ValueError("target_style must be 'matched' or 'title_matched'.")

    records.append({'input': input_text, 'target': target_text})


# Save to JSONL
with open(output_jsonl_path, 'w') as f:
    for r in records:
        json.dump(r, f)
        f.write('\n')

# Output confirmation
print(f"JSONL saved to: {output_jsonl_path}")
print("Sample entry:")
print(json.dumps(records[0], indent=2))
```
